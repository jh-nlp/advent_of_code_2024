```python
import pandas as pd
```

```python
df = pd.read_csv('input_2.txt', header = None)
df.columns = ['values']
df['values'] = df['values'].str.split()
df
```

```python
def str_to_int(lst):
    return [int(x) for x in lst]
```

```python
df['values'] = df['values'].apply(str_to_int)
df
```

```python
def safe_logic(lst: list)-> bool:
    diffs =  [right- left for left, right in zip(lst[1:], lst[:-1])]
    positive = diffs[0] >= 0 
    for diff in diffs:
        if (diff >= 0) != positive:
            return False
        if abs(diff) < 1 or abs(diff) > 3:
            return False
    return True
    

safe_logic([8,6 ,4, 4, 1])        
```

```python
df['safe'] = df['values'].apply(safe_logic)
df
```

```python
def fuzz_safe(lst:list[int]):
    print(lst)
    if safe_logic(lst) is False:
        naughty_counter = 0
        for index, current_val in enumerate(lst):
            rest = lst[:index] + lst[index+1:]
            print(rest)
            if safe_logic(rest) is True:
                naughty_counter += 1
                # print(f'remove {i} is fine')
            else:
                # print(f'remove {i} is still not fine')
                pass
        if naughty_counter >= 1:
            return True    
        else:
            return False

    else: 
        return False
```

```python
df['fuzzy_safe'] = df['values'].apply(fuzz_safe)
df
```

```python
df['overall'] = df.apply(lambda row: (row['safe'] is True) or (row['fuzzy_safe'] is True), axis = 1 )
```
