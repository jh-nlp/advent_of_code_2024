```python
found_mul = re.finditer(r'mul\(\d+,\d+\)', s)
muls= dict()
for i, match_obj in enumerate(found_mul):
    muls.update({match_obj.span()[0]:match_obj.group()})
muls
```

```python
found_dos = re.finditer(r'do\(\)', s)
dos = dict()
for match_obj in found_dos:
    dos.update({match_obj.span()[0]:True})
```

```python
found_donts = re.finditer(r'don\'t\(\)', s)
donts = dict()
for match_obj in found_donts:
    donts.update({match_obj.span()[0]:False})
```

```python
combined = {**muls, **dos , **donts}
combined = dict(sorted(combined.items()))
```

```python
def multiple_literals(mul_str:str):
    pair_str = re.findall(r'\d+,\d+', mul_str)[0]
    lst = [int(num) for num in  pair_str.split(',')]
    if (lst[0]> 0 and lst[0]<=999) and (lst[1]> 0 and lst[1]<=999):
        mul = lst[0]* lst[1]
    return mul
```

```python
do = True
result = 0
for i, act in combined.items():
    if isinstance(act, str):
        if do:
            result+= multiple_literals(act)
    elif isinstance(act, bool):
        do = act

result     
```
